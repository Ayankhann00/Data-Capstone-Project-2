```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('netflix.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df['premiere']=pd.to_datetime(df['premiere'])
```

```python
df['premiere'].head()
```

```python
df['title'].unique()
```

```python
df.info()
```

```python
df['premiere']=df['premiere'].dt.year
```

```python
df['premiere'].head()
```

```python
df.head()
```

Sucessfully converted premiere object to date time and to integer values 


```python
cols=['year']
```

```python
df.drop(cols,axis=1,inplace=True)
```

```python
df.head()
```

```python

df['duration_category'] = df['runtime'].apply(lambda x: 'Short' if x < 80 else 'Medium' if x <= 100 else 'Long')
```

```python
df.head()
```

```python
df['rating_score']=df['imdb_score'].apply(lambda x: 'Average' if x==6.0 else 'Poor' if x<6.0 else 'Good'if x>6.0 and x<8.0 else 'Worth Watching')
```

```python
df.head(50)
```

```python
df['score_runtime_product'] = df['imdb_score'] * df['runtime']
```

```python
df.head()
```

```python
df['rating_score'].value_counts()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df['genre']=df['genre'].str.split('/')
```

```python
df=df.explode('genre').reset_index(drop=True)
```

```python
df.head(10)
```

```python
df['genre']=df['genre'].astype('category')
```

```python
df['genre'].dtypes
```

```python
df.info()
```

```python
df.nunique()
```

```python
sns.set_style('darkgrid')
```

```python
df['genre'].describe()
```

```python
g=sns.catplot(y='genre',data=df,kind='count',order=df['genre'].value_counts().index,height=13,aspect=0.5,color='purple')
g.set_axis_labels("Count", "Genre")  # Label axes
plt.title("Distribution of Music Genres", pad=20)  # Add title with padding
plt.tight_layout()  # Prevent overlapping elements
plt.show()
```

```python
plt.figure(figsize=(10, 16)) 

g = sns.stripplot(
    x='imdb_score',
    y='genre',
    data=df,
    hue='language',
    palette='coolwarm',
    jitter=True,
    alpha=0.6
)

plt.xlabel("IMDb Score")
plt.ylabel("Genre")
plt.title("Distribution of IMDb Scores by Genre and Language", pad=20)
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  
plt.show()
```

```python
plt.figure(figsize=(8, 6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='magma', linewidths=0.5, linecolor='black')
plt.title('Correlation Heatmap of Netflix Dataset')
plt.show()
```

```python
sns.histplot(df['runtime'],kde=True,bins=30)
```

```python
plt.figure(figsize=(14, 12))
sns.barplot(
    y='genre',
    x='imdb_score',
    data=df,
    hue='duration_category',
    estimator=np.std,  
    palette='coolwarm'
)

plt.title('Standard Deviation of IMDb Scores by Genre and Language', fontsize=19)
plt.xlabel('IMDb Score (Standard Deviation)')
plt.ylabel('Genre')
plt.legend(title='Category', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
sns.lmplot(
    x='premiere',
    y='imdb_score',
    data=df,
    aspect=1.5,
    height=6,
    line_kws={'color': 'red'},
    scatter_kws={'alpha': 0.5}
)

plt.title("Linear Regression: IMDb Score vs Premiere Year", fontsize=14)
plt.xlabel("Premiere Year")
plt.ylabel("IMDb Score")
plt.tight_layout()
plt.show()
```

```python
df['premiere'].hist()
plt.title('Release Year')
plt.xlabel('Premiere')
plt.tight_layout()
```

```python
df.groupby('premiere')['imdb_score'].mean().plot()
```

```python
# Top 10 genres by frequency
top_genres = df['genre'].value_counts().nlargest(10).index

# Filter the dataset
filtered_df = df[df['genre'].isin(top_genres)]

# Plot
plt.figure(figsize=(17, 13))  # Increased height to prevent overlap
sns.boxplot(y='genre', x='imdb_score', data=filtered_df, palette='viridis')

plt.title('IMDb Score Distribution for Top 10 Genres')
plt.xlabel('IMDb Score')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```
